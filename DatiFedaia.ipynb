#### Vengono tolte manualmente le colonne non interessate

```python
for i in range(2,27,2):
    data.pop('Unnamed: '+str(i))
for i in range(1,7,2):
    data.pop('Unnamed: '+str(i))
for i in range(17,27,2):
    data.pop('Unnamed: '+str(i))
```

```python
#data
```

```python
data=data.rename(columns={'Unnamed: 0':'Anno','Unnamed: 7':'1h','Unnamed: 9':'3h','Unnamed: 11':'6h','Unnamed: 13':'12h','Unnamed: 15':'24h'})
```

```python
data=data.iloc[3:,:]
```

```python
#data
```

```python
data=data.set_index('Anno')
data
```

#### Conversione in dati numerici di tutti i valori contenuti nel file excel

```python
lst=['1h','3h','6h','12h','24h']
for i in lst:
    data[i]=pd.to_numeric(data[i], errors='coerce')
```
